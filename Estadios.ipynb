# 3. Criando Mapa


Latitude e longitude central

```python
center_lat = -14.24007
center_long = -53.18050
```

Colocando a lista de estádios em uma variável

```python
consulta = list(collection.find())

consulta_pd = pd.DataFrame(consulta)
```

Mapa

```python
mapa = folium.Map(location=[center_lat, center_long], zoom_start=4, width=800, height=600)

qtde_pontos = len(consulta_pd)
```

Criando marcadores

```python
for i in range(qtde_pontos):
  lng = float(consulta_pd.iloc[i]["longitude"])
  lat = float(consulta_pd.iloc[i]["latitude"])
  descricao = "Estádio: " + str(consulta_pd.iloc[i]["nome"]) + "<br> Capacidade: " + str(consulta_pd.iloc[i]["capacidade"])
  folium.Marker([lat, lng], popup=descricao, icon=folium.Icon("red")).add_to(mapa)
```
